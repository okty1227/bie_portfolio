```python
import pandas as pd
df = pd.read_csv('preprocess_project2.csv')
```

```python
df.loc[df['Year']==2019,:].groupby(['State']).sum()['count(Severity)'].sort_values(ascending=False)[:5].to_json('2019.json')
```

```python
df.loc[df['Year']==2020,:].groupby(['State']).sum()['count(Severity)'].sort_values(ascending=False)[:5].to_json('2020.json')
```
