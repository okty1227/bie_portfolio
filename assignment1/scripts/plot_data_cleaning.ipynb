```python
import pandas as pd
```

```python
data = pd.read_csv('youtube.csv', encoding='unicode_escape')
data.head(10)
```

```python
data.category.unique()
```

```python
data.channel_type.unique()
```

```python
data.columns
```

```python
us_data = data[(data.Abbreviation == 'US')]
us_data = us_data[['subscribers', 'subscribers_for_last_30_days', 'video views', 'video_views_for_the_last_30_days', \
                   'category', 'channel_type', 'lowest_monthly_earnings', 'highest_monthly_earnings', \
                   'lowest_yearly_earnings','highest_yearly_earnings']]
us_data
```

```python
us_data.isna().sum()
```

```python
us_data.rename({'subscribers_for_last_30_days': 'subscribers_30_days',\
                'video views': 'views',\
                'video_views_for_the_last_30_days': 'views_30_days'}, axis=1, inplace=True)
us_data
```

```python
us_data.channel_type.value_counts()
```

## Category

```python
us_cat_filtered = us_data.groupby(['category']).filter(lambda cat: len(cat) > 5)
view_sub_ratio = (us_cat_filtered.views / us_cat_filtered.subscribers).mean()
us_cat = us_cat_filtered.groupby(['category']).mean()
us_cat_max = us_cat_filtered.groupby(['category']).max()
us_cat['max_yearly_earnings'] = us_cat_max.highest_yearly_earnings
us_cat['ratio'] = view_sub_ratio
us_cat.reset_index(inplace=True)
us_cat
# us_cat.to_csv('youtube_us_cat9.csv', encoding='utf-8', index=False)
```

```python
us_cat = us_data.groupby(['category']).mean()
us_cat_max = us_data.groupby(['category']).max()
us_cat['max_yearly_earnings'] = us_cat_max.highest_yearly_earnings
us_cat.reset_index(inplace=True)
us_cat
# us_cat.to_csv('youtube_us_cat.csv', encoding='utf-8', index=False)
```

## Type

```python
us_type_filtered = us_data.groupby(['channel_type']).filter(lambda t: len(t) > 5)
view_sub_ratio = (us_type_filtered.views / us_type_filtered.subscribers).mean()
us_type = us_type_filtered.groupby(['channel_type']).mean()
us_type_max = us_type_filtered.groupby(['channel_type']).max()
us_type['max_yearly_earnings'] = us_type_max.highest_yearly_earnings
us_type['ratio'] = view_sub_ratio
us_type.reset_index(inplace=True)
us_type
# us_type.to_csv('youtube_us_type9.csv', encoding='utf-8', index=False)
```

```python
us_type = us_data.groupby(['channel_type']).mean()
us_type_max = us_data.groupby(['channel_type']).max()
us_type['max_yearly_earnings'] = us_type_max.highest_yearly_earnings
us_type.reset_index(inplace=True)
us_type
# us_type.to_csv('youtube_us_type.csv', encoding='utf-8', index=False)
```
